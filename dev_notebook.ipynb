```python
#Module usefull for extraction data
from modules.LoggerManager import LoggerManager
import pandas as pd
from modules.MarketDataManager import MarketDataManager
from modules.AnalyzerTools import AnalyzerTools
from modules.DisplayTools import DisplayTools
pd.set_option('display.max_rows', 500)

analyze = AnalyzerTools()
display = DisplayTools()

#Create Market and download data from config
myExchange = MarketDataManager(exchange_name="binance")
await myExchange.download_data(['SOL/USDT'],['1h'],"2023-10-01 00:00:00")

init = myExchange.load_data_from_csv_db(coin_pair="SOL-USDT", interval_name='1h', start_date="2023-10-01 00:00:00", end_date="2024-01-31 00:00:00")
top_bot = analyze._trend_get_top_botom(ohlv=init, candle_min_window=3)
top_bot['bottom'].replace(1, '^', inplace=True)
top_bot['top'].replace(1, 'v', inplace=True)
top = top_bot['top'].to_list()
bottom = top_bot['bottom'].to_list()
mark =[]
for i in range(len(top)):
    if top[i] != 0:
        mark.append(top[i])
    elif bottom[i] != 0 :
        mark.append(bottom[i])
    else : 
        mark.append("")

len(mark)
```

```python
range_volatility = 10
range_min_max = 5
top_bottom_reader = 3
finding_window = 20

# Calcul des indicateurs de volatilit
init["volatility"] = abs(init["close"] - init["open"]).rolling(range_volatility).mean()
init["min_close"] = init["close"].rolling(range_min_max).min()
init["max_close"] = init["close"].rolling(range_min_max).max()
init["unique"] = range(len(init))
        
ohlv = analyze._trend_get_top_botom(ohlv=init, candle_min_window=top_bottom_reader)
ohlv['volatility'] = ohlv['volatility'].fillna(0)
lst_index_top = ohlv.loc[(ohlv["top"] == 1) & (ohlv["volatility"] != 0)]['unique'].to_list()
lst_index_bottom = ohlv.loc[(ohlv["bottom"] == 1) & (ohlv["volatility"] != 0)]['unique'].to_list()
```

```python
rate_max_local = 0.20
rate_max_volatility = 0.1

double_bot = []
print(lst_index_bottom)

for bot in lst_index_bottom:
    i=1
    finder=0
    flag_max_local = False
    print("--------------------------------------------")
    while (i <= finding_window + 1 and finder < len(init)):
        finder = bot + i
        # print(f'Finder = {finder} | statut max_local : {flag_max_local}')
        if ohlv.iloc[bot]["close"] + rate_max_local * ohlv.iloc[bot]["volatility"] < ohlv.iloc[finder]["close"]:
            # print(f'{ohlv.iloc[bot]["close"] + rate_max_local * ohlv.iloc[bot]["volatility"]} || {ohlv.iloc[finder]["close"]}')           
            flag_max_local=True
        if flag_max_local and ohlv.iloc[bot]["close"] - ohlv.iloc[finder]["close"] < rate_max_volatility * ohlv.iloc[finder]["volatility"]:
            max_local_price = ohlv.iloc[bot:finder]["close"].max()
            max_local_index = ohlv.iloc[bot:finder][ohlv["close"] == max_local_price]["unique"].values[0]
            bot_init_index = bot
            bot_find_index = finder
            bot_av_price = round((ohlv.iloc[bot]["close"] + ohlv.iloc[bot]["close"])/2,2)
            # print(f'max_local_index : {max_local_index} bot_init_index : {bot_init_index} bot_find_index : {bot_find_index} max_local_price : {max_local_price} bot_av_price : {bot_av_price}')
            double_bot.append({
                "max_local_index" : max_local_index,
                "bot_init_index" : bot_init_index,
                "bot_find_index" : bot_find_index,
                "max_local_price" : max_local_price,
                "bot_av_price" : bot_av_price,                
            })
        i += 1
```
